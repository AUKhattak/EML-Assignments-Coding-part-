### 3. Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#linear_model = LinearRegression()

y = np.array(data['mpg'])
x = np.array(data['cylinders'])

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=0)
print("shape",y_test.shape)
print(y_test)

print(x.shape, y.shape)
# Use the fit function and the score function in linear regression module of sklearn to  fit the data and observe the fit
#newData=data.drop(['origin', 'weight', 'name', 'year'], axis=1)

linear_model = LinearRegression().fit(X_train.reshape(-1,1),y_train)
print("fitness score",linear_model.score(X_train.reshape(-1,1),y_train))
y_predict=linear_model.predict(X_test.reshape(-1,1))
#print(linear_model.get_params())
r2score=r2_score(y_test, y_predict)
print("R2 score", r2score)
```

```python
# Fit the linear regression on mpg for the other features: cylinder, horsepower, year and displacement
y = np.array(data['mpg'])
x_displacement = np.array(data['displacement'])

# linear_model = LinearRegression().fit(x_displacement.reshape(-1,1), y)
# print("displacement vs. mpg: ", linear_model.score(x_displacement.reshape(-1, 1), y))

X_train, X_test, y_train, y_test = train_test_split(x_displacement, y, test_size=0.33, random_state=0)
print("shape",y_test.shape)
print(y_test)


linear_model = LinearRegression().fit(X_train.reshape(-1,1),y_train)
print("fitness score",linear_model.score(X_train.reshape(-1,1),y_train))
y_predict=linear_model.predict(X_test.reshape(-1,1))
#print(linear_model.get_params())
r2score=r2_score(y_test, y_predict)
print("R2 score", r2score)

```

```python
# Fit the linear regression on mpg for the other features: cylinder, horsepower, year and displacement

y = np.array(data['mpg'])
x_horsepower = np.array(data['horsepower'])


# linear_model = LinearRegression().fit(x_displacement.reshape(-1,1), y)
# print("displacement vs. mpg: ", linear_model.score(x_displacement.reshape(-1, 1), y))



X_train, X_test, y_train, y_test = train_test_split(x_horsepower, y, test_size=0.33, random_state=0)
print("shape",y_test.shape)
print(y_test)


linear_model = LinearRegression().fit(X_train.reshape(-1,1),y_train)
print("fitness score",linear_model.score(X_train.reshape(-1,1),y_train))
y_predict=linear_model.predict(X_test.reshape(-1,1))
#print(linear_model.get_params())
r2score=r2_score(y_test, y_predict)
print("R2 score", r2score)

```

```python
# Fit the linear regression on mpg for the other features: cylinder, horsepower, year and displacement

y = np.array(data['mpg'])
x_year = np.array(data['year'])



X_train, X_test, y_train, y_test = train_test_split(x_year, y, test_size=0.33, random_state=0)
print("shape",y_test.shape)
print(y_test)


linear_model = LinearRegression().fit(X_train.reshape(-1,1),y_train)
print("fitness score",linear_model.score(X_train.reshape(-1,1),y_train))
y_predict=linear_model.predict(X_test.reshape(-1,1))
#print(linear_model.get_params())
r2score=r2_score(y_test, y_predict)
print("R2 score", r2score)
```

```python
#answer 5.3:
# the predictor that has highest significant relationship with outcome is "displacement" as we can see the R-squared value
# for this model is highst among all. Resulting models of mpg with year is not very good model, rest of three models are 
# satisfactory based on Rsqaured value. cylinders, displacement and horsepower has significant relationship to the outcome.
```

```python
#Multiple Linear Regression
#Create a 2D matrix with each column representing all features
#Look up np.concatenate or np.stack


y = np.array(data['mpg'])

x1 = np.array(data['cylinders'])
x2 = np.array(data['displacement'])
x3 = np.array(data['horsepower'])
x4 = np.array(data['weight'])
x5 = np.array(data['year'])
x6 = np.array(data['origin'])
x7 = np.array(data['acceleration'])

data_set = np.stack(( x1, x2, x3, x4, x5, x6, x7), axis = 1)
print(data_set.shape)
#print(data_set)
X_train, X_test, y_train, y_test = train_test_split(data_set, y, test_size=0.33, random_state=0)


linear_model = LinearRegression().fit(X_train,y_train) 
print("all predictors together versus mpg: SCORE : ", linear_model.score(X_train,y_train))

# #parameters = linear_model.get_params()
# #parameters
y_predict=linear_model.predict(X_test)
#print(y_predict)
#print(y_test)
# ans

#Print the predictions of the your fitted model and the weights of Linear Regression (look at the function get_params())
#print("Weights : ", linear_model.coef_)
i=1
for weight in linear_model.coef_:
    print("w",i,": ",weight)
    i = i+1

print("intercept : ", linear_model.intercept_)
linear_model.get_params(deep=True)
r2=r2_score(y_test, y_predict)
print("R2 value: ", r2)
```

```python
# answer 5.4
# the models that we have generated in 5.3 has less Rsquare value than the model generated in 5.4 (combibed), that shows this
# model is best fit and can make better predictions.
# in previous model when we mapped individual predictor with mpg we can see that relationship is not very good but when we used 
# all these variables together and mapped against mpg we can see a strong relationship and a good regression model.
# the sign value of coeefecient tell us the kind of relationship exist between them, if the sign is positive that means both 
# are changing in same direction, otherwise opposite direction.
# in our fitted modle the coefficient value of horsepower and weight (w3 and w4) the value is negative that means they are 
# changing in opposite directions
```
